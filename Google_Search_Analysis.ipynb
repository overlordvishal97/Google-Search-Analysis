```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
Trending_topics = TrendReq(hl='en-US', tz=360)
```

**Build Payload**

```python
kw_list =['Cloud Computing']
Trending_topics.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='', gprop='')
```

**Interest Over Time**

```python
Trending_topics.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='', gprop='')
data = Trending_topics.interest_over_time()
data = data.sort_values('Cloud Computing', ascending=False)
data.head(10)
print(data)
```

**Interest By Region**

```python
data = Trending_topics.interest_by_region()
data = data.sort_values('Cloud Computing', ascending=False)
data.head(10)
print(data)
```

**Bar chart**

```python
data.reset_index().plot(x='geoName', y='Cloud Computing', figsize=(20,10), kind='bar')
plt.style.use('fivethirtyeight')
plt.title('Cloud Computing')
plt.show()
```

**Top charts**

```python
df = Trending_topics.top_charts(2020, hl='en-US', tz=300, geo='GLOBAL')
df.head(10)
```

**Related Queries**

```python

from pytrends.request import TrendReq

# Initialize the TrendReq object (avoid overwriting it)
pytrends_obj = TrendReq()
pytrends_obj.build_payload(kw_list=['Cloud Computing'])

# There's no direct way to access the raw response in the current pytrends version.
# Consider raising an issue on the pytrends GitHub if you need this functionality.

# Instead, you can try to extract related queries directly
related_queries = pytrends_obj.related_queries()
if related_queries:
    print(related_queries.values())
else:
    print("No related queries found.")
```
